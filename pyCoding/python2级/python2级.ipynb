```python

cc =0

l=[]
def permutation(ln,start,end):
    global cc
    if start==end:
        print(ln)
        l.append(str(ln))
        
        cc = cc+1
    else:
        for i in range(start,end+1):
            ln[start],ln[i]  =  ln[i],ln[start]
            permutation(ln,start+1,end)
            ln[start],ln[i]  =  ln[i],ln[start]
            
            
num=eval(input('请输入一个list'))
permutation(num,0,len(num)-1)
print(f'cc is {cc}')

s=set(l)
print(len(l))
print(len(s))  
for i in l:
    print(i)
  
for i in s:
    print(i)
            
                
```

```python
def fact(n):
    if n==1:
        return 1
    else:
        return n*fact(n-1)
a=eval(input('请输入一个整数：'))
print(fact(a)) 

print(2**64)
print(fact(64))   
```

```python
def fei(n):
    if n==0:
        # print( 0)
        return 0
    elif n==1:
        # print( 1 )
        return 1
    else:
        # print( fei(n-1)+fei(n-2))
        return fei(n-1)+fei(n-2)
    
def fei_loop(n):
    
    """
    [0,1,1,2,3, ..... ,f(n)]
    """
    l = []
    for i in range(n+1):
        if i<=1:
            l.append(i)
        else:
            l.append(l[i-1]+l[i-2])        
    return l[-1] 
 
a=eval(input('请输入一个整数：'))
for i in range(a+1):
    print(f"{i}  : "  ,  fei_loop(i))
```

```python
#求素数
import math
primeList = []
def getPrime(n):
    for x in range(2, n+1):
        x_sqrt =math.sqrt(x)
        for prime in primeList:
            if prime>x_sqrt:
                primeList.append(x)
                break
            if x%prime==0:
                break
        else:
            primeList.append(x)
    return primeList
print(getPrime(9))


#交作业
def getHomeWork(hw, n):
    if n == len(hw)-1:
        return hw[n]
    return getHomeWork(hw,n+1)+hw[n]
homew = ["作业1", "作业2", "作业3", "作业4", "作业5", "作业6", "作业7"]

print(homew[::-1])
zuoye = getHomeWork(homew, 0)
print(zuoye)

#排列组合
COUNT = 0
def perm(n, begin, end):
    global COUNT
    if begin>= end:
        print(n)
        COUNT += 1
    else:
        i=begin
        for num in range(begin, end):
            n[num], n[i] = n[i], n[num]
            perm(n, begin+1, end)
            n[num], n[i] = n[i], n[num]
arr = input("1-n的第1个排列：")
n = [int(n) for n in arr.split()]
perm(n, 0, len(n))
print("排列总数是：%d个！" % COUNT)
```

```python
#爬楼梯，每次可爬1，2节
def climbStairs__loop( n: int) -> int:
        if n <= 2:
            return n
        n1 = 1
        n2 = 2
        for i in range(3, n+1):
            temp = n1+n2
            n1, n2 = n2, temp
        return temp

# @cache
def climbStairs(n: int) -> int:
    if n <= 2:
        return n
    else:
        return climbStairs(n-1) + climbStairs(n-2)
            
            
print(climbStairs(5))   
print(climbStairs__loop(5))
```

```python
import turtle as t
t.TurtleScreen._RUNNING = True

#基础设置
t.pensize(5)
t.speed(0)
t.bgcolor('red')

#简单背景
t.penup()
t.goto(0, -300)
t.fillcolor('white')
t.begin_fill()
t.circle(300, extent=360)
t.end_fill()
t.goto(0,320)
t.write('奥利给将军',font=('楷体',70),align='center')
t.goto(-300, 100)
t.fillcolor('yellow')
t.begin_fill()
for i in range(5):
    t.fd(600)
    t.right(144)
t.end_fill()
t.goto(100,50)
t.pendown()

def yanjing(x,y,gao=50, kuan=25, bangao=40, tong=10):
    t.penup()
    t.goto(x, y)
    t.pendown()
    t.seth(0)
    t.fillcolor('white')
    t.begin_fill()
    t.left(90)
    t.fd(gao)
    t.circle(kuan, extent=180)
    t.fd(gao)
    t.end_fill()
    t.penup()
    t.goto(x-kuan,y+gao*2/3)
    t.pendown()
    t.dot(tong,'black')
    t.penup()
    
def huzi(x, y, huzi_kuan, huzi_gao):
    t.penup()
    t.goto(x, y)
    t.pendown()
    t.seth(0)
    #胡子ooooooooooooooo
    t.fillcolor('black')
    t.begin_fill()
    for i in range(2):
        t.fd(huzi_kuan)
        t.right(90)
        t.fd(huzi_gao)
        t.right(90)
    t.end_fill()

def zui(x, y,r):
    t.penup()
    t.goto(x, y)
    t.pendown()
    t.seth(0)
    t.right(30)
    t.circle( r, extent=60)
    
def maozi(x, y):
    t.penup()
    t.goto(x, y)
    t.pendown()
    t.seth(0)
    #帽子oooooooooooooooooooooooooooooooooooooooooo
    t.left(90)
    t.fillcolor('blue')
    t.begin_fill()
    for i in range(3):
        t.fd(60)
        t.left(90)
        t.fd(40)
        t.left(90)
    t.end_fill()
    ##帽身lllllllllllllllllllllllllllllllll
    t.fillcolor('orange')
    t.begin_fill()
    t.right(135)
    t.fd(40)
    t.right(135)
    t.fd(96.56)
    t.right(135)
    t.fd(40)
    t.right(180)
    t.end_fill()

    #帽顶llllllllllllllllllllllllllllllllll
    t.fd(15)
    t.left(-45)
    t.fd(10)
    t.circle(22.07, extent=180)
    t.fd(40.6)
    t.left(90)
    t.penup()
    t.fd(10)
    t.pendown()

    t.pencolor('red')
    t.right(90)
    t.fd(10)
    t.right(180)
    for i in range(5):
        t.fd(20)
        t.right(144)
    t.fd(10)
    t.right(-90)
    t.penup()
    t.pencolor('black')
    t.fd(10)
    t.pendown()
    t.left(90)
    t.fd(40.6)
    t.circle(22.07, extent=180)
    t.fd(10)

def lian(x, y, gao, kuan):
    t.penup()
    t.goto(x, y)
    t.pendown()
    t.seth(0)
    t.fillcolor('lightgreen')
    t.begin_fill()
    for i in range(2):
        t.circle(gao, extent=180)
        t.fd(kuan)
    t.end_fill()    

def qiang():
    
    t.fillcolor('gray')
    t.begin_fill()
    t.right(90)
    t.fd(25)

    t.left(90)
    t.fd(30)
    t.right(90)
    t.fd(30)
    t.fd(-30)
    t.left(90)
    t.fd(70)
    t.left(90)
    t.fd(20)
    t.left(90)
    t.fd(70)
    t.right(90)
    t.fd(40)
    t.left(90)
    t.fd(30)
    t.left(90)
    t.fd(35)
    t.end_fill()
    t.right(180)
    t.fd(35)
    t.left(-90)
    t.fd(30)
    t.left(-90)
    t.fd(40)
    t.left(90)
    t.fd(20)
    t.left(90)
    t.circle(20, extent=90)
    t.right(90)
    t.fd(20)
    t.left(90)
    t.fd(30)
    t.left(90)
    t.fd(35)
    t.left(90)

def shenzi(x, y, jiangao=30,jiaodu_left=135, jiaodu_right=135, bi=120, shen=130, yanse='lightgreen', shenkuan=100, tuichang=120,
    yanse2='green', shencha=140, shenkuan2=40, dugao=130,draw_qiang=1): 
        #    shenzi(15, 45, 135, 60, 75, '', 50, 60, '', 70, 20, 65)


    t.penup()
    t.goto(x, y)
    t.pendown()
    t.seth(0)

    t.right(90)
    t.fd(jiangao)
    t.right(jiaodu_left)
    t.fd(bi)
    if draw_qiang:
        qiang()

    t.fd(-bi)
    t.left(jiaodu_left)
    t.right(180)
    t.fd(jiangao)
    t.right(180)
    t.fillcolor(yanse)
    t.begin_fill()
    t.fd(shen)
    t.circle(shenkuan, extent=60)
    t.right(60)
    t.fd(tuichang)
    t.right(180)
    t.fd(tuichang)
    t.right(120)
    t.circle(shenkuan, extent=60)
    t.right(120)
    t.fd(tuichang)
    t.right(180)
    t.fd(tuichang)
    t.right(60)
    t.circle(shenkuan, extent=60)
    t.fd(shen-jiangao)
    t.right(jiaodu_right)
    t.fd(bi)
    t.fd(-bi)
    t.left(jiaodu_right)
    t.fd(jiangao)
    #lllllllllllllllllllllllllllllllllllllllllllll
    t.end_fill()
    t.fillcolor(yanse2)
    t.begin_fill()
    t.left(90)
    t.fd(shencha)
    t.left(90)
    t.fd(dugao)
    t.circle(shenkuan2, extent=180)
    t.fd(dugao)
    t.end_fill()
    t.left(90)

def pifeng(x,y,p_chang,p_jiao,lian_kuan):
    t.penup()
    t.goto(x, y)
    t.pendown()
    t.seth(0)

    t.fillcolor('red')
    t.begin_fill()
    t.right(p_jiao)
    t.fd(p_chang)
    t.left(35)
    a = ((lian_kuan/8)*(lian_kuan/8)/2)**0.5
    for i in range(8):
        t.fd(a)
        t.left(90)
        t.fd(a)
        t.left(-90)

    t.goto(100, 50)
    t.left(45)
    t.end_fill()

def qiangjia(x,y,x_end,y_end):
    t.goto(x,y)
    t.left(140)
    t.fd(200)
    t.left(40)
    t.fillcolor('brown')
    t.begin_fill()
    t.fd(50)
    t.right(90)
    t.fd(70)
    t.right(90)
    t.fd(30)
    t.right(90)
    t.fd(45)
    t.right(-90)
    t.fd(20)
    t.right(90)
    t.fd(25)
    t.end_fill()
    t.right(90)
    t.fd(40)
    t.left(90)
    t.goto(x_end, y_end)

def draw1(x,y):
    lian(x,y,20,100)
    yan_x = x
    yan_y = y+40
    yanjing(yan_x, yan_y, gao=25, kuan=12.5, bangao=20, tong=5)
    zui(x-75,y+20,50)
    yan_x = yan_x-75
    yan_y = yan_y
    yanjing(yan_x, yan_y, gao=25, kuan=12.5, bangao=20, tong=5)
    t.right(90)
    shenzi(x-100,y,15,45,135,60,75,'lightgreen',50,60,'green',70,20,75,0)

def draw_shi(x,y):
    t.penup()
    t.goto(x, y)
    t.pendown()
    t.seth(0)
    t.fillcolor('brown')
    t.begin_fill()
    t.left(180)
    for i in range(3):
        t.right(120)
        t.fd(60)
    for d in range (3):
        for i in range(2):
            t.circle(20,extent=180)
            t.fd(60+d*40)
        t.circle(20, extent=180)
        t.right(180)
        t.fd(20)
    t.end_fill()

def xing(x,y,yanse,chang):
    t.penup()
    t.goto(x, y)
    t.pendown()
    t.seth(0)
    t.penup()
    t.fillcolor(yanse)
    t.begin_fill()
    for i in range(5):
        t.fd(chang)
        t.right(144)
    t.end_fill()

def draw_xiao(x, y):
    t.penup()
    t.goto(x, y)
    t.pendown()
    t.seth(0)
    yanjing(x,y,40,20,20,10)
    yanjing(x-75, y, 40, 20, 20, 10)
    t.goto(x-80,y-20)
    t.seth(0)
    t.pendown()
    t.right(90)
    t.circle(25,extent=180)
    t.penup()
    t.goto(x-80, y-20)
    t.pendown()
    t.right(180)
    t.fd(120)
    t.circle(25, extent=180)
    t.right(-90)
    t.fd(50)
    t.fd(-50)
    t.left(-90)
    t.fd(120)

def draw_su(x,y):
    t.penup()
    t.goto(x, y)
    t.seth(0)
    t.penup()
    xing(x,y,'yellow',80)
    t.fillcolor('yellow')    
    t.left(25)
    t.goto(x+10,y-100)
    t.begin_fill()
    for i in range(2):
        t.fd(40)
        t.left(90)
        t.fd(20)
        t.left(90)
    t.fd(25)
    for i in range(2):
        t.right(90)
        t.fd(120)
        t.right(90)
        t.fd(15)
    t.end_fill()
    t.right(90)
    t.fd(35)
    t.right(90)
    t.fd(55)
    t.begin_fill()
    for i in range(2):
        t.fd(40)
        t.left(90)
        t.fd(20)
        t.left(90)
    t.end_fill()        
    t.left(90)
    t.width(15)
    t.pendown()
    t.pencolor('yellow')
    t.circle(55,extent=170)
    t.circle(90, extent=60)
    t.pencolor('black')
    t.width(5)

def draw_aoligei(x, y, gao, kuan):
    t.penup()
    t.goto(x, y)
    t.pendown()
    t.seth(0)
    lian(x, y, gao, kuan)
    pifeng(x-kuan,y,230,80,kuan)
    #计算右眼睛的位置
    yan_x = x
    yan_y = y+gao*2
    yanjing(yan_x, yan_y, gao=50, kuan=25, bangao=40, tong=10)
    #计算左眼睛的位置
    yan_x = yan_x-kuan+50
    yan_y = yan_y
    yanjing(yan_x, yan_y, gao=50, kuan=25, bangao=40, tong=10)
    #画胡子
    huzi_kuan = 40
    huzi_gao = 20
    huzi_x = x-kuan/2-huzi_kuan/2
    huzi_y = y+gao*2-10
    huzi(huzi_x, huzi_y, huzi_kuan, huzi_gao)
    #画嘴
    zui_r = huzi_kuan/2*10
    zui_x = x-kuan/2-zui_r/2
    zui_y = y+gao*2-10-huzi_gao
    zui(zui_x, zui_y, zui_r)
    t.speed(0)
    #画帽子
    maozi_x = x-kuan/2+huzi_kuan/2
    maozi_y = y+2*gao
    maozi(maozi_x, maozi_y)
    #画身子
    shen_x = x-kuan
    shen_y = y
    shenzi(shen_x, shen_y)
    #画枪夹
    qiangjia_x=shen_x
    qiangjia_y=y
    qiangjia(qiangjia_x,qiangjia_y,x,y)

draw1(430, 350)
draw_shi(-400,-300)
draw_xiao(430,-250)
draw_su(-430,400)
xing(370,0,'yellow',144)
xing(-470, 0, 'yellow',144)
draw_aoligei(100, 50, 30, 200)
t.done()
```
